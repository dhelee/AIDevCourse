## 1. 표본분포

```python
# 표본추출방법 - 랜덤추출

import random
[random.randint(1, 1000) for i in range(10)]
```

### 표본평균의 분포

```python
import numpy as np

# 표준정규분포를 따르는 모집단에서 n=10인 표본을 10000번 추출하여 그 표본평균을 구함 
xbars = [np.mean(np.random.normal(size=10)) for i in range(10000)]
print("mean %f, var %f" %(np.mean(xbars), np.var(xbars)))

# 평균이 0, 분산이 1/10에 가깝게 나오는 것을 확인할 수 있다
```

```python
# normal(loc=평균, scale=표준편차, size=표본크기)
xbars = [np.mean(np.random.normal(loc=10, scale=3, size=10)) for i in range(10000)]
print("mean %f, var %f" %(np.mean(xbars), np.var(xbars)))

# 히스토그램 그리기
import matplotlib.pyplot as plt
h = plt.hist(xbars, range=(5, 15), bins=30)
```

### 중심극한정리

```python
n=3
# random.rand(n)은 0~1사이의 랜덤수를 n개 추출
# np.random.rand(3) * 10 -> 0~10사이의 랜덤수를 3개 선택
# 모집단은 평균이 5, 분산이 (10 - 0)^2 / 12 = 2.89인 uniform distribution
xbars = [np.mean(np.random.rand(n) * 10) for i in range(10000)]
print("mean %f, var %f" %(np.mean(xbars), np.var(xbars)))
h = plt.hist(xbars, range=(0, 10), bins=100)
```

```python
# n의 크기를 늘려보자

n=10
xbars = [np.mean(np.random.rand(n) * 10) for i in range(10000)]
print("mean %f, var %f" %(np.mean(xbars), np.var(xbars)))
h = plt.hist(xbars, range=(0, 10), bins=100)
```

```python
# n의 크기를 늘려보자

n=30
xbars = [np.mean(np.random.rand(n) * 10) for i in range(10000)]
print("mean %f, var %f" %(np.mean(xbars), np.var(xbars)))
h = plt.hist(xbars, range=(0, 10), bins=100)
```

```python
# 지수분포도 확인해보자
# exponential(scale=평균, size=표본크기)
n=2
xbars = [np.mean(np.random.exponential(scale=3, size=n)) for i in range(10000)]
print("mean %f, var %f" %(np.mean(xbars), np.var(xbars)))
h = plt.hist(xbars, range=(0, 10), bins=100)
```

```python
# n의 크기를 늘려보자

n=30
xbars = [np.mean(np.random.exponential(scale=3, size=n)) for i in range(10000)]
print("mean %f, var %f" %(np.mean(xbars), np.var(xbars)))
h = plt.hist(xbars, range=(0, 10), bins=100)
```

## 추정

### 모평균의 구간추정

```python
w = [10.7, 11.7, 9.8, 11.4, 10.8, 9.9, 10.1, 8.8, 12.2, 11.0, 11.3, 11.1, 10.3, 10.0, 9.9, 11.1, 11.7, 11.5, 9.1, 10.3, 8.6, 12.1, 10.0, 13.0, 9.2, 9.8, 9.3, 9.4, 9.6, 9.2]
xbar = np.mean(w)
sd = np.std(w, ddof=1)
print("평균 %.2f, 표준편차 %.2f" %(xbar, sd))
```

```python
import scipy.stats

# 95% 신뢰구간
alpha = 0.05
# norm.ppf -> 임계값 Z(a/2)를 반환하는 함수
zalpha = scipy.stats.norm.ppf(1-alpha/2)
print("zalpha: ", zalpha)
```

### 모비율의 구간추정

```python
x=48
n=150
phat=x/n
alpha=0.05
zalpha = scipy.stats.norm.ppf(1-alpha/2)
sd=np.sqrt(phat*(1-phat)/n)
print("phat: %.3f, zalpha: %.3f, sd: %.3f"%(phat, zalpha, sd))
ci=[phat - zalpha * sd, phat + zalpha * sd]
print(ci)
```

## 모평균의 검정

```python
w = [10.7, 11.7, 9.8, 11.4, 10.8, 9.9, 10.1, 8.8, 12.2, 11.0, 11.3, 11.1, 10.3, 10.0, 9.9, 11.1, 11.7, 11.5, 9.1, 10.3, 8.6, 12.1, 10.0, 13.0, 9.2, 9.8, 9.3, 9.4, 9.6, 9.2]
mu = 10.5
xbar = np.mean(w)
sd = np.std(w, ddof=1)
print("평균 %.2f, 표준편차 %.2f" %(xbar, sd))

z = (xbar-mu)/(sd/np.sqrt(len(w)))
print("검정통계량:", z)

alpha = 0.05
cri = scipy.stats.norm.ppf(1-alpha/2)
print("임계값:", cri)

# 검정통계량이 임계값보다 작으므로 귀무가설을 기각할 수 없다
```

## 교차엔트로피


```python
def crossentropy(P, Q):
    return sum([-P[i]*np.log2(Q[i]) for i in range(len(P))])

P = [1,0]
Q = [0.8, 0.2]
print(crossentropy(P,Q))
Q = [0.5, 0.5]
print(crossentropy(P,Q))
Q = [0.2, 0.8]
print(crossentropy(P,Q))
```
