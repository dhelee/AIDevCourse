## 1. Matplotlib 시작하기
- 파이썬의 데이터 시각화 라이브러리

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

## 2. 자주 사용되는 Plotting의 Options
- `plt.plot()`: 꺾은선 그래프를 그리는 함수

```python
# .plot()의 인자로 전달한 리스트가 y값, 각 원소의 인덱스가 x값
plt.plot([1, 2, 3, 4, 5]) # 실제 plottion을 하는 함수 # y = x + 1 그래프
plt.show() # plt를 확인하는 명령
```

- `plt.figure()`: plotting을 할 도면을 선언
- `figsize`: figure()의 인자로 도면(그래프) 크기 조절

```python
plt.figure(figsize=(6, 6))

plt.plot([0, 1, 2, 3, 4]) # y = x 
plt.show()
```

### 2차함수 그리기 with plot()

```python
# numpy.array()를 이용해서 함수 그래프 그리기
# y = x^2 그리기

x = np.array([1, 2, 3, 4, 5]) # 정의역
y = np.array([1, 4, 9, 16, 25]) # f(x)

plt.plot(x, y)
plt.show()
```

```python
# np.arange(a, b, c) c: 0.01 
# python의 range는 c로 integer만 가능, numpy의 arrange는 c로 실수 가능

x = np.arange(-10, 10, 0.01)
plt.plot(x, x**2)
plt.show()
```

- `plt.xlabel()`, `plt.ylabel()`: x, y축에 설명 추가하기

```python
x = np.arange(-10, 10, 0.01)

###
plt.xlabel("x value")
plt.ylabel("f(x) value")
###

plt.plot(x, x**2)
plt.show()
```

- `plt.axis()`: x, y축의 범위 설정하기

```python
x = np.arange(-10, 10, 0.01)

plt.xlabel("x value")
plt.ylabel("f(x) value")

###
plt.axis([-5, 5, 0, 25]) # [x_min, x_max, y_min, y_max]
###

plt.plot(x, x**2)
plt.show()
```

- `plt.xticks()`, `plt.yticks()`: x, y축에 눈금 설정하기

```python
x = np.arange(-10, 10, 0.01)

plt.xlabel("x value")
plt.ylabel("f(x) value")

plt.axis([-5, 5, 0, 25]) 

###
plt.xticks([i for i in range(-5, 6, 1)]) # x축 눈금 설정
plt.yticks([i for i in range(0, 27, 3)]) # y축 눈금 설정
###

plt.plot(x, x**2)
plt.show()
```

- `plt.title()`: 그래프에 title 달기

```python
x = np.arange(-10, 10, 0.01)

plt.xlabel("x value")
plt.ylabel("f(x) value")

plt.axis([-5, 5, 0, 25]) 

plt.xticks([i for i in range(-5, 6, 1)]) 
plt.yticks([i for i in range(0, 27, 3)]) 

###
plt.title("y = x^2 graph")
###

plt.plot(x, x**2)
plt.show()
```

- `plt.legend()`: 범례 설정하기

```python
x = np.arange(-10, 10, 0.01)

plt.xlabel("x value")
plt.ylabel("f(x) value")

plt.axis([-5, 5, 0, 25]) 

plt.xticks([i for i in range(-5, 6, 1)]) 
plt.yticks([i for i in range(0, 27, 3)]) 

plt.title("y = x^2 graph")

###
plt.plot(x, x**2, label="trend")
plt.legend()
###

plt.show()
```

## 3. Matplotlib Case Study

### 꺾은선 그래프(Plot)
- `plt.plot()`
- 시계열 데이터에 많이 사용

```python
x = np.arange(20) # 0~19
y = np.random.randint(0, 20, 20) # 난수를 20번 생성

plt.plot(x, y)

# y축을 20까지 보이게 하고 싶다면?
plt.axis([0, 20, 0, 20])
# y축을 5단위로 보이게 하고 싶다면?
plt.yticks([0, 5, 10, 15, 20]) # y축 눈금 설정

plt.show()
```

### 산점도(Scatter Plot)
- `plt.scatter()`
- x와 y사이의 상관관계를 파악하는 데 사용

```python
plt.scatter(x, y)
plt.show()
```

### 상자 그림(Box Plot)
- `plt.boxplot()`
- 수치형 데이터에 대한 정보(Q1, Q2, Q3, min, max)

```python
plt.boxplot((x, y))

# plot title 지정하기
plt.title("Box plot of x, y")

plt.show()
```

### 막대 그래프(Bar Plot)
- `plt.bar()`
- 범주형 데이터의 "값"과 그 값의 크기를 직사각형으로 나타낸 그림

```python
plt.bar(x, y)

plt.xticks(np.arange(0, 20, 1))

plt.show()
```

```python
# cf) Histogram
# 도수분포를 직사각형의 막대 형태로 나타낸다.
# "계급"으로 나타낸 것이 특징: 변량을 그룹화하여 "범주형" 데이터로 구성

plt.hist(y, bins=np.arange(0, 22, 2))

plt.xticks(np.arange(0, 22, 2))

plt.show()
```

### 원형 그래프(Pie Chart)
- `plt.pie()`
- 데이터에서 전체에 대한 부분의 비율을 부채꼴로 나타낸 그래프
- 다른 그래프에 비해서 **비율** 확인에 용이

```python
z = [100, 300, 200, 400]

plt.pie(z, labels=['one', 'two', 'three', 'four'])
plt.show()
```

## 4. The 멋진 그래프, Seaborn Case Study
Matplotlib를 기반으로 더 다양한 시각화 방법을 사용하는 라이브러리

```python
import seaborn as sns
```

### 커널밀도그림(Kernel Density Plot)
- 히스토그램과 같은 연속적인 분포를 곡선화해서 그린 그림
- `sns.kdeplot()`

```python
# in histogram

x = np.arange(0, 22, 2)
y = np.random.randint(0, 20, 20)

plt.hist(y, bins=x)
plt.show()
```

```python
import seaborn as sns

# kdeplot
# y가 density를 나타냄

sns.kdeplot(y, shade=True)

plt.show()
```

### 카운트그림(Count Plot)
- 범주형 column의 빈도수를 시작화 -> Groupby 후의 도수를 하는 것과 동일한 효과
- `sns.countplot()`

```python
vote_df = pd.DataFrame({'name':['Andy', 'Bob', 'Cat'], "vote":[True, True, False]})

vote_df
```

```python
# in matplotlib

vote_count = vote_df.groupby('vote').count()

vote_count
```

```python
plt.bar(x=[False, True], height=vote_count['name'])
```

```python
# countplot

sns.countplot(x=vote_df['vote'])
```

### 캣그림(Cat Plot)
- 숫자형 변수와 하나 이상의 범주형 변수의 관계를 보여주는 함수
- `sns.catplot()`

```python
covid = pd.read_csv('dataset/country_wise_latest.csv')

covid.head(5)
```

```python
# strip plot
s = sns.catplot(x='WHO Region', y='Confirmed', data=covid)
s.fig.set_size_inches(10, 6)
```

```python
# violin plot
s = sns.catplot(x='WHO Region', y='Confirmed', data=covid, kind='violin')
s.fig.set_size_inches(10, 6)
```

### 스트립그림(Strip Plot)
- scatter plot과 유사하게 데이터의 수치를 표현하는 그래프
- `sns.stripplot()`

```python
sns.stripplot(x="WHO Region", y='Recovered', data=covid)
```

```python
# cf) swarmplot
# 점들이 겹치는 경우 옆으로 분산해서 표현

sns.swarmplot(x="WHO Region", y='Recovered', data=covid)
```

### 히트맵(Heatmap)
- 데이터의 행렬을 색상으로 표현해주는 그래프
- 상관계수 그래프를 그리는 데 주로 사용
- `sns.heatmap()`

```python
covid.corr()
```

```python
sns.heatmap(covid.corr())
```
